```python
import torch
from diffusers import StableDiffusionPipeline
import PIL.Image
import os

# Function to generate images
def generate_images(prompt="a serene sunset over a futuristic city", num_images=3, output_folder="generated_images"):
    os.makedirs(output_folder, exist_ok=True)

    model_id = "runwayml/stable-diffusion-v1-5"
    pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
    pipe.to("cuda" if torch.cuda.is_available() else "cpu")

    images = []
    for i in range(num_images):
        image = pipe(prompt).images[0]
        file_path = os.path.join(output_folder, f"image_{i+1}.png")
        image.save(file_path)
        images.append(file_path)
        print(f"Saved {file_path}")

    return images

# Generate images
generated_images = generate_images()
```

```python
import cv2
import numpy as np
import torchvision.transforms as transforms

# Function to preprocess images
def preprocess_image(image_path, output_folder="preprocessed_images"):
    os.makedirs(output_folder, exist_ok=True)

    image = cv2.imread(image_path)
    image = cv2.resize(image, (224, 224))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = image / 255.0  # Normalize to [0, 1]

    tensor_transform = transforms.ToTensor()
    image_tensor = tensor_transform(image)

    output_path = os.path.join(output_folder, os.path.basename(image_path))
    cv2.imwrite(output_path, (image * 255).astype(np.uint8))
    print(f"Preprocessed and saved {output_path}")

    return image_tensor

# Process all generated images
processed_tensors = [preprocess_image(img) for img in generated_images]
```
